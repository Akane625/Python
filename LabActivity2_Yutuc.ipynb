Four different classes related to your program
Define at least 3 properties for each class
Create at least 3 methods (One method will display the current values of the properties.  The other two is your choice)
1 class will use a constructor to initialize
2 classes will not use a constructor but assign values to the properties or attributes

```python
class Info:
  def __init__(self, name, age, id):
    self.name = name
    self.age = age
    self.id = id
  def advance(self):
    self.age += 4
  def display(self):
    print("Name: ", self.name)
    print("Age: ", self.age)
    print("ID: ", self.id)

current = Info("Thom", 18, 104232)
current.advance()
current.display()
```

```python
class Program:
  course = "Data Science and Analytics"
  subject = "Intermediate Programming"
  topic = "Object Oriented Programming"
  def update(self, new_topic):
    self.topic = new_topic
  def change(self, new_subject):
    self.subject = new_subject
  def show(self):
    print("Course: ", self.course)
    print("Subject: ", self.subject)
    print("Topic: ", self.topic)

topic = Program()
topic.update("Recursion")
topic.change("Data Structures and Algorithms")
topic.show()
```

```python
class Grades:
  THY2 = 90
  MC4153 = 100
  CC4152 = 95
  def mean(self):
    average = (self.THY2 + self.MC4153 + self.CC4152) / 3
    return average
  def grade(self):
    print("Semestral Average: ", self.mean())
  def present(self):
    print("Theology 2: ", self.THY2)
    print("Calculus 2: ", self.MC4153)
    print("Intermediate Programming: ", self.CC4152)

number = Grades()
number.present()
number.grade()
```

```python
class Statistics:
  weight_1 = 54
  weight_2 = 74
  weight_3 = 35
  def mean(self):
    average = (self.weight_1 + self.weight_2 + self.weight_3) / 3
    return average
  def meanshow(self):
    print(f"Average weight: {self.mean():.2f}")
  def change(self, new_weight):
    self.weight_3 = new_weight
  def exhibit(self):
    print("Patient 1: ", self.weight_1)
    print("Patient 2: ", self.weight_2)
    print("Patient 3: ", self.weight_3)

item = Statistics()
item.change(66)
item.exhibit()
item.meanshow()
```
