Write a code that takes two words and return the longest word and the shortest word and length of each.(Note: User input)

```python
word_1 = input("Enter first word: ")
word_2 = input("Enter second word: ")

if len(word_1) > len(word_2):
  print(f"The longest word is {word_1} ({len(word_1)} characters) and the shortest word is {word_2} ({len(word_2)} characters)")
else:
  print(f"The longest word is {word_2} ({len(word_2)} characters) and the shortest word is {word_1} ({len(word_1)} characters)")
```

Write a code to change a given string to a newly string where the first and last chars have been exchanged. (Note: User input)

```python
string = input("Enter a word: ")
rep_start = input("Enter replacement (beginning): ")
rep_end = input("Enter replacement (end): ")

new = rep_start + string[1:-1] + rep_end

print(new)
```

Write a code that will ask the user to input a string and display separately the vowels, consonants and other characters(special characters e.g. '#', '?', etc.)

```python
thing = list(input("Enter a string: "))

vowel = "aeiouAEIOU"
consonant = "qwrtypsdfghjklzxcvbnmQWRTYPSDFGHJKLZXCVBNM"

n_vow = []
n_con = []
n_var = []

for i in thing:
  if i in vowel:
    n_vow.append(i)
  elif i in consonant:
    n_con.append(i)
  else:
    n_var.append(i)

print("Vowels: ", end = "")
for i in n_vow:
  print(i, end = " ")

print("\nConsonants: ", end = "")
for i in n_con:
  print(i, end = " ")

print("\nOther characters: ", end = "")
for i in n_var:
  print(i, end = " ")
```
