```python
# print outs the file. you can add .head() to print the first few lines

import pandas as pd

df = pd.read_csv(r"C:\Users\Thom\Python Projects\Jupyter\Iris Dataset\a08a1080b88344b0c8a7-0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534\iris.csv")

print(df)
```

```python
# extract the numbers of a certain column

numeric_sepal_length = pd.to_numeric(df['sepal_length'], errors = 'coerce').dropna()

print(numeric_sepal_length)
```

```python
# calculate any math/stats for all columns without numpy

species = df.drop(columns = ['species'])
species_average = df.groupby('species').mean()
print(species_average)

# calculate any math/stats for a column without numpy

means_petal_length = df.groupby('species')['petal_length'].mean()
print()
print(means_petal_length)
```

```python
# finding the mean

import numpy as np

data = df.iloc[:, :-1].values
means = np.mean(data, axis = 0)

print(means)

# if you want the labels

for label, total in zip(df.columns[:-1], means):
    print(f"{label}: {total}")
```

```python
# calculate the median

data = df.iloc[:, :-1].values
medians = np.median(data, axis = 0)

print(medians)

# if you want labels

for label, total in zip(df.columns[:-1], medians):
    print(f"{label}: {total}")
```

```python
# calculate the mode

data = df.iloc[:, :-1].values

modes = []
for thing in data.T:
    unique, counts = np.unique(thing, return_counts = True)
    mode = unique[np.argmax(counts)]
    modes.append(mode)

print(modes)

# if you want labels

for label, total in zip(df.columns[:-1], modes):
    print(f"{label}: {total}")
```

```python
# calculate the mode (alterntive)
from scipy import stats

data = df.iloc[:, :-1].values
modes = stats.mode(data, axis = 0, keepdims = False)

print(modes.mode)

# if you want labels

for label, total in zip(df.columns[:-1], modes.mode):
    print(f"{label}: {total}")
```

```python
# calculate the sum

data = df.iloc[:, :-1].values
sums = np.sum(data, axis = 0)

print(sums)

# if you want the labels

for label, total in zip(df.columns[:-1], sums):
    print(f"{label}: {total}")
```

```python
# calculate the cumulative sum (all)

data = df.iloc[:, :-1].values
cumulative_sums = np.cumsum(data, axis = 0)

print(cumulative_sums)

# if you want labels

for i, label in enumerate(df.columns[:-1]):
    print(f"{label}: ")
    print(cumulative_sums[:, i])
    print()
```

```python
# Graph 1 (Line Graph)

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Thom\Python Projects\Jupyter\Iris Dataset\a08a1080b88344b0c8a7-0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534\iris.csv")

x1, y1 = df['sepal_width'], df['sepal_length']
x2, y2 = df['petal_width'], df['petal_length']

fig, (ax1, ax2, ay1, ay2) = plt.subplots(4, 1, figsize = (15, 30))
ax1.plot(x1, color = "red")
ax2.plot(x2, color = "green")
ay1.plot(y1, color = "yellow")
ay2.plot(y2, color = "purple")

ax1.set_title("Sepal Width")
ay1.set_title("Sepal Length")
ax2.set_title("Petal Width")
ay2.set_title("Petal Length")

ax1.set_xlabel("Amount")
ax1.set_ylabel("Size")
ax2.set_xlabel("Amount")
ax2.set_ylabel("Size")
ay1.set_xlabel("Amount")
ay1.set_ylabel("Size")
ay2.set_xlabel("Amount")
ay2.set_ylabel("Size")

plt.show()
```

```python
# Graph 2 (Scatter Plot)

fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10, 15))
ax1.scatter(x1, y1, color = "red")
ax2.scatter(x2, y2, color = "green")

ax1.set_title("Sepal Length and Width")
ax2.set_title("Petal Length and Width")

ax1.set_xlabel("Width")
ax1.set_ylabel("Length")
ax2.set_xlabel("Width")
ax2.set_ylabel("Length")

plt.show()
```

```python
# Graph 3 (Bar Graph)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 10))
ax1.bar(x1, y1, color = "red")
ax2.bar(x2, y2, color = "green")

ax1.set_title("Sepal Length and Width")
ax2.set_title("Petal Length and Width")

ax1.set_xlabel("Width")
ax1.set_ylabel("Length")
ax2.set_xlabel("Width")
ax2.set_ylabel("Length")

plt.show()
```

```python
# Graph 6 (Pie Chart)

fig, ((ax1, ax2), (ay1, ay2)) = plt.subplots(2,2, figsize = (10, 10))
ax1.pie(x1)
ax2.pie(x2)
ay1.pie(y1)
ay2.pie(y2)

ax1.set_title("Sepal Width")
ay1.set_title("Sepal Length")
ax2.set_title("Petal Width")
ay2.set_title("Petal Length")

plt.show()
```

```python
# Graph 7 (Histogram)

fig, ((ax1, ax2), (ay1, ay2)) = plt.subplots(2,2, figsize = (10, 10))
ax1.hist(x1, bins = 30, color = "red")
ax2.hist(x2, bins = 30, color = "green")
ay1.hist(y1, bins = 30, color = "yellow")
ay2.hist(y2, bins = 30, color = "purple")

ax1.set_title("Sepal Width")
ay1.set_title("Sepal Length")
ax2.set_title("Petal Width")
ay2.set_title("Petal Length")

ax1.set_xlabel("Size")
ax1.set_ylabel("Amount")
ax2.set_xlabel("Size")
ax2.set_ylabel("Amount")
ay1.set_xlabel("Size")
ay1.set_ylabel("Amount")
ay2.set_xlabel("Size")
ay2.set_ylabel("Amount")

plt.show()
```

```python
# Graph 4 (2D Array)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

Z1 = np.sin(np.sqrt(x ** 2 + y ** 2)) #tested different equations for a 2d structure
Z2 = np.cos(np.sqrt(x ** 2 + y ** 2))
Z3 = np.sin(x) * np.cos(y)
Z4 = np.exp(-0.1 * (x ** 2 + y ** 2))

fig, ((ax1, ax2), (ay1, ay2)) = plt.subplots(2,2, figsize = (10, 10))

ax1.imshow(Z1, extent = [-10, 10, -10, 10], origin = 'lower', cmap = 'viridis')
ax2.imshow(Z2, extent = [-10, 10, -10, 10], origin = 'lower', cmap = 'plasma')
ay1.imshow(Z3, extent = [-10, 10, -10, 10], origin = 'lower', cmap = 'inferno')
ay2.imshow(Z4, extent = [-10, 10, -10, 10], origin = 'lower', cmap = 'cividis')

ax1.set_title("Sepal Width")
ay1.set_title("Sepal Length")
ax2.set_title("Petal Width")
ay2.set_title("Petal Length")

plt.show()
```

```python
# Graph 5 (Contour Plot)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

Z1 = np.sin(np.sqrt(x ** 2 + y ** 2)) #tested different equations for a 2d structure
Z2 = np.cos(np.sqrt(x ** 2 + y ** 2))
Z3 = np.sin(x) * np.cos(y)
Z4 = np.exp(-0.1 * (x ** 2 + y ** 2))

fig, ((ax1, ax2), (ay1, ay2)) = plt.subplots(2,2, figsize = (10, 10))

ax1.contourf(x, y, Z1, cmap = "Reds")
ax2.contourf(x, y, Z2, cmap = "Greens")
ay1.contourf(x, y, Z3, cmap = "Blues")
ay2.contourf(x, y, Z4, cmap = "Purples")

ax1.set_title("Sepal Width")
ay1.set_title("Sepal Length")
ax2.set_title("Petal Width")
ay2.set_title("Petal Length")

plt.show()
```
