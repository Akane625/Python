```python
# A

from math import pi
from abc import ABC, abstractmethod

class shape(ABC):
    def __init__(self):
        self.area = 0
    @abstractmethod
    def calculateArea(self):
        pass
    def display(self):
        print(f"Area is {self.area}")

class rectangle(shape):
    def calculateArea(self, l, w):
        self.area = l * w

class circle(shape):
    def calculateArea(self, r):
        self.area = pi * (r ** 2)

class triangle(shape):
    def calculateArea(self, b, h):
        self.area = (1 / 2) * b * h

rect = rectangle()
rect.calculateArea(2, 4)

circ = circle()
circ.calculateArea(4)

tria = triangle()
tria.calculateArea(3, 5)

shapes = [rect, circ, tria]
for shape in shapes:
    shape.display()
```

```python
# B

class shape:
    def __init__(self,name, length, width):
        self.length = length
        self._width = width
        self.__name = name
    def info(self):
        print(f"Item: {self.__name} Length: {self.length} Width: {self._width}")
    def get_name(self):
        return self.__name

class rectangle(shape):
    def __init__(self, name, length, width):
        super().__init__(name, length, width)
    def get_area(self):
        area = self.length * self._width
        print(f"Area: {area}")
    def display(self):
        print(f"This is a {self.get_name()}")
        
thing = shape("Cardboard Box", 2, 5)
thing.info()

rect = rectangle("Cardboard Box", 2, 5)
rect.get_area()
rect.display()
```
